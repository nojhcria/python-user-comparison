```python
# ==========================================================
# Importação da biblioteca Pandas responsável pela leitura, manipulação e análise dos dados.
# ==========================================================

import pandas as pd

# ==========================================================
# Leitura dos arquivos CSV exportados dos dois ambientes FAC.
# Cada arquivo representa a lista de usuários de um ambiente.
# ==========================================================

ctc = pd.read_csv("CTC_users.csv")
dtc = pd.read_csv("DTC_users.csv")


# ==========================================================
# Exibe as primeiras linhas de cada arquivo para validar se a importação foi realizada corretamente.
# ==========================================================

#display(ctc.head())
#display(dtc.head())
```

```python
# ==========================================================
# Exibe a quantidade total de registros (usuários) encontrados em cada ambiente.
# ==========================================================

print("Quantidade de usuários no CTC:", len(ctc))
print("Quantidade de usuários no DTC:", len(dtc))
```

```python
# ==========================================================
# Exibe o nome das colunas presentes em cada arquivo.
# Esta etapa foi utilizada para identificar o nome correto da coluna que contém o login dos usuários.
# ==========================================================

ctc.columns
dtc.columns
```

```python
# ==========================================================
# Cria um conjunto (set) contendo apenas os usernames.
#
# O uso de "set" elimina possíveis usuários duplicados e permite realizar comparações de forma muito mais rápida.
# ==========================================================

usuarios_ctc = set(ctc["# username"])
usuarios_dtc = set(dtc["# username"])

# ==========================================================
# Exibe a quantidade total de usuários únicos encontrados em cada ambiente.
# ==========================================================

print("Total CTC:", len(usuarios_ctc))
print("Total DTC:", len(usuarios_dtc))

```

```python
# ==========================================================
# Identifica usuários existentes no ambiente CTC, mas ausentes no ambiente DTC.
#
# A operação de diferença entre conjuntos retorna apenas os elementos exclusivos do primeiro conjunto.
# ==========================================================

somente_ctc = usuarios_ctc - usuarios_dtc

#somente_ctc
```

```python
# ==========================================================
# Identifica usuários existentes no ambiente DTC, mas ausentes no ambiente CTC.
# ==========================================================

somente_dtc = usuarios_dtc - usuarios_ctc

#somente_dtc
```

```python
# ==========================================================
# Exibe a quantidade de divergências encontradas em cada ambiente.
# ==========================================================

print("Usuários somente no CTC:", len(somente_ctc))
print("Usuários somente no DTC:", len(somente_dtc))
```

```python
# ==========================================================
# Cria uma lista vazia que será utilizada para montar o relatório final contendo todas as divergências.
# ==========================================================

divergencias = []

# ==========================================================
# Percorre todos os usuários encontrados apenas no CTC e adiciona essas informações na lista do relatório.
# ==========================================================

for usuario in somente_ctc:

    divergencias.append({

        "Usuario": usuario,

        "Origem": "CTC",

        "Situacao": "Existe no CTC, mas não existe no DTC"

    })


# ==========================================================
# Percorre todos os usuários encontrados apenas no DTC e adiciona essas informações na lista do relatório.
# ==========================================================

for usuario in somente_dtc:

    divergencias.append({

        "Usuario": usuario,

        "Origem": "DTC",

        "Situacao": "Existe no DTC, mas não existe no CTC"

    })


# ==========================================================
# Converte a lista de dicionários em um DataFrame.
#
# O DataFrame facilita a visualização dos dados e também permite exportá-los para outros formatos.
# ==========================================================

df_divergencias = pd.DataFrame(divergencias)

df_divergencias


# ==========================================================
# Exporta o relatório final para um arquivo Excel.
#
# O parâmetro index=False impede que o índice numérico seja gravado como uma coluna no arquivo.
# ==========================================================

df_divergencias.to_excel("Relatorio_Divergencias_FAC.xlsx",

    index=False

)

```
